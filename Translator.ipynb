```python
# Replace the filename below with your dataset's actual name
data = pd.read_csv('Language Detection.csv')   # e.g. "language_data.csv"
print("✅ Dataset loaded successfully!")
print(data.head())

# ensure required columns exist (checking for both cases)
assert ('text' in data.columns and 'language' in data.columns) or ('Text' in data.columns and 'Language' in data.columns), "Dataset must have 'text' and 'language' columns"

# Rename columns to lowercase if they are capitalized
if 'Text' in data.columns:
    data = data.rename(columns={'Text': 'text', 'Language': 'language'})


# clean up
data = data.dropna(subset=['text', 'language'])
print(f"Total usable samples: {len(data)}")
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['language'], test_size=0.2, random_state=42)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # You can adjust max_features

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Transform the testing data
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print("✅ Data vectorized successfully!")
print(f"Shape of training data (TF-IDF): {X_train_tfidf.shape}")
print(f"Shape of testing data (TF-IDF): {X_test_tfidf.shape}")

# Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=1000) # Increase max_iter if convergence warning appears
model.fit(X_train_tfidf, y_train)

print("✅ Model trained successfully!")

# Make predictions on the test set
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)


def predict_language(text):
    # Vectorize the input text using the trained TF-IDF vectorizer
    text_tfidf = tfidf_vectorizer.transform([text])

    # Predict the language
    predicted_language = model.predict(text_tfidf)[0]

    return predicted_language

# Example usage:
new_text = "This is a test sentence."
predicted_language = predict_language(new_text)
print(f"The predicted language of '{new_text}' is: {predicted_language}")

new_text_french = "Bonjour, comment ça va?"
predicted_language_french = predict_language(new_text_french)
print(f"The predicted language of '{new_text_french}' is: {predicted_language_french}")

new_text_spanish = "Hola, ¿cómo estás?"
predicted_language_spanish = predict_language(new_text_spanish)
print(f"The predicted language of '{new_text_spanish}' is: {predicted_language_spanish}")
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['language'], test_size=0.2, random_state=42)

print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # You can adjust max_features

# Fit and transform the training data
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)

# Transform the testing data
X_test_tfidf = tfidf_vectorizer.transform(X_test)

print("✅ Data vectorized successfully!")
print(f"Shape of training data (TF-IDF): {X_train_tfidf.shape}")
print(f"Shape of testing data (TF-IDF): {X_test_tfidf.shape}")

# Initialize and train the Logistic Regression model
model = LogisticRegression(max_iter=1000) # Increase max_iter if convergence warning appears
model.fit(X_train_tfidf, y_train)

print("✅ Model trained successfully!")

# Make predictions on the test set
y_pred = model.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(report)


def detect_language(text):
    vec = tfidf_vectorizer.transform([text])
    pred_code = model.predict(vec)[0]
    confidence = np.max(model.predict_proba(vec))
    # A basic lang_map, you can expand this with more language codes and names
    lang_map = {
        'en': 'English', 'fr': 'French', 'es': 'Spanish', 'de': 'German', 'it': 'Italian',
        'pt': 'Portuguese', 'ru': 'Russian', 'ar': 'Arabic', 'zh': 'Chinese', 'ja': 'Japanese',
        'ko': 'Korean', 'hi': 'Hindi', 'ml': 'Malayalam', 'ta': 'Tamil', 'kn': 'Kannada',
        'tr': 'Turkish', 'nl': 'Dutch', 'da': 'Danish', 'sv': 'Swedish', 'el': 'Greek'
    }
    lang_name = lang_map.get(pred_code, pred_code)
    print(f"\n🗣️ Detected Language: {lang_name} ({pred_code})")
    print(f"🎯 Confidence: {confidence*100:.2f}%")

while True:
    txt = input("\nEnter text (or type 'exit' to quit): ")
    if txt.lower() == 'exit':
        break
    detect_language(txt)
```
